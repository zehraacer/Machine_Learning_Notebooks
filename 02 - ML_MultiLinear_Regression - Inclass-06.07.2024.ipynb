# Train-Test Split

```python
# ML için kullanılan kütüphane Scikit-Learn kütüphanesidir

# !pip install scikit-learn 
# !pip install sklearn
```

```python
X = df.drop(columns="sales") # df[["TV", "radio", "newspaper"]] , Indep.variable'lar X e atandı
y = df["sales"]   #Modelimin önüne hedef olarak target variable atadım.
print(X, y)
```

```python
from sklearn.model_selection import train_test_split

#Veri setinizi randomly olarak eğitim (training) ve test (testing) alt kümelerine ayırmak için kullanılır.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

X_train: Eğitim veri setinin özelliklerini (features) içerir.

X_test: Test veri setinin özelliklerini (features) içerir.

y_train: Eğitim veri setinin etiketlerini (labels/targets) içerir.

y_test: Test veri setinin etiketlerini (labels/targets) içerir.

train_test_split(X, y, test_size=0.3, random_state=42): Bu fonksiyon çağrısı ile X (özellikler) ve y (etiketler) veri setleri rastgele olarak eğitim ve test alt kümelerine ayrılır.

# Model Fitting and Compare Actual and Predicted Labels

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()

#variable a atanmaz ise ilerleyen satırlarda prediction yaparken hata alınabiliyor
```

```python
model.fit(X_train, y_train) #tarin data seti üzerinde eğitim gerçekleşiyor
```

```python
model.coef_

 # modelin katsayılarını verir (TV:0.04, radio:0.199, newsp:0.006)
```

```python
model.feature_names_in_  

# yukardak katsayıların karşılık geldiği featurelerı verir
```

```python
model.intercept_

#modeldeki sabit  sayıyı verir
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = b1*TV + b2*radio + b3*newspaper + b0
```

```python
# herhangi bir satırdaki değerleri tahmin edelim. manuel olarak y_pred hesabı yapalım
# bu deger y_test kumesindeki 95.satır oaln 16.9 gercek degerin manuel olarak tahmin hesabı yapılmış hali olacaktır

sum(X_test.loc[95]*model.coef_) + model.intercept_
```

```python
#actual: gerçek değerler   #pred: tahmin edilen değerler    #residual: gerçek değer - tahmin değeri

my_dict = {"Actual": y_test, "pred": y_pred, "residual":y_test - y_pred}
```

```python
comparing = pd.DataFrame(my_dict)
comparing
```

```python
#comparing isimli dataframe den örnek olarak 25 tane çekelim

result_sample = comparing.head(25)
```

```python
result_sample
```

# Error Metrics

```python
#modelin başarısını ölçelim
```

```python
# performans değerlkendirme için kullanacağımız error metrikleri
# RMSE degerini manuel hesaplayacağız

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
R2_score = r2_score(y_test, y_pred)
R2_score
```

```python
# MAE hesabı  (ne kadar düşük o kadar iyi)
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
# mean squared error hesabı

mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# Root Mean Squared Error - RMSE  hesabı
    
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
#Ortalama Hata Ornaı
```

```python
sales_mean= df["sales"].mean()
sales_mean
```

```python
#MAE Hata oranı - Sıfıra yakın olması iyi bir model olduğunu gösterir(NMAE)

mae / sales_mean
```
